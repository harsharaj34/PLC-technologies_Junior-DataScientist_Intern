```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and axis
fig, ax = plt.subplots()

# Plotting the data with customizations
ax.plot(x, y1, label='Sine Wave', color='blue', linestyle='-', linewidth=2, marker='o', markersize=5)
ax.plot(x, y2, label='Cosine Wave', color='red', linestyle='--', linewidth=2, marker='x', markersize=5)

# Adding titles and labels
ax.set_title('Customized Sine and Cosine Waves', fontsize=16, fontweight='bold')
ax.set_xlabel('X-axis', fontsize=14)
ax.set_ylabel('Y-axis', fontsize=14)

# Adding grid lines
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Customizing ticks
ax.set_xticks(np.arange(0, 11, 1))
ax.set_yticks(np.arange(-1, 1.5, 0.5))

# Adding legend
ax.legend(loc='upper right', fontsize=12)

# Adding annotations
ax.annotate('Max of Sine', xy=(1.57, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Customize spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)

# Display the plot
plt.show()
```

```python
#seaborn
#Customizing a Histogram
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Customizing a histogram
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Customizing a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time', style='time', s=100, palette='viridis')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend(title='Time of Day')
plt.grid(True)
plt.show()
```

```python
# Customizing a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill', palette='Set2')
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.grid(True)
plt.show()
```

```python
# Customizing a pair plot
sns.pairplot(tips, hue='sex', palette='coolwarm', markers=['o', 's'], plot_kws={'alpha': 0.5})
plt.suptitle('Pair Plot of Tips Dataset', y=1.02)
plt.show()
```
